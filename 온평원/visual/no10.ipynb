```python
import pandas as pd
result = pd.read_csv('/Users/zoostar/Downloads/download_file_10_1.csv',encoding='cp949')
data = {}
for i in range(1,len(result)):
    data[result.iloc[i,0]]= result.iloc[1, range(i,len(result.columns),6)]

name = input('세종특별자치시의 행정구역을 입력해주세요: ') 
for i,j in data.items():
    if i.find(name) != -1:
        for k in range(len(data[i])):
            data[i][k] = int(data[i][k].replace(',',''))
            plt.bar(data[i].index[k].split('년_')[0],data[i][k])
        plt.ylim([min(data[i])-3000,max(data[i])+3000])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

result = pd.read_csv('/Users/zoostar/Downloads/download_file_10_2.csv',encoding='cp949')
x_label=[]
y_label=[]
for i in range(len(result.columns)):
    if result.columns[i].startswith('2020년_거주자'):
        x_label.append(result.columns[i])
        if type(result.iloc[0,i]) == str:
            result.iloc[0,i] = int(result.iloc[0,i].replace(',',''))
        y_label.append(result.iloc[0,i])
```

```python
plt.pie(y_label[2:],labels=x_label[2:])
plt.xticks(rotation=45)
plt.show()
```
