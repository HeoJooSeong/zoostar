```python
import csv
f = open('/Users/zoostar/Downloads/download_file.csv','r',encoding='cp949')
result = csv.reader(f)
next(result)
next(result)
data = {}
for row in result:
    data[row[0]] = []
    for i in range(1,len(row),6):
        data[row[0]].append(row[i])
print(data)
```
