This first maps a line to an integer value, creating a new RDD. reduce is called on that RDD to find the largest line count. The arguments to map and reduce are Python anonymous functions (lambdas), but we can also pass any top-level Python function we want. For example, weâ€™ll define a max function to make this code easier to understand:

```python
def max(a, b):
...     if a > b:
...         return a
...     else:
...         return b
...
```

```python
from stop_words import get_stop_words
def stripNonAlphaNum(text):
    import re
    return re.compile(r'\W+', re.UNICODE).split(text)
```

```python
def clr(tmp):
    tmp = tmp.replace(',),','')
    tmp = tmp.replace(',','')
    tmp = tmp.replace(',)','')
    tmp = tmp.replace('(','')
    tmp = tmp.replace('\"','')
    tmp = tmp.replace('/',' ')
    tmp = tmp.replace(".\'",'')
    tmp = tmp.replace("u'",'')
    tmp = tmp.lower()
    return tmp
```

```python
result = str(df.filter(df.id>0).collect())
stop_words = get_stop_words('en')
text = clr(result)
text = ' '.join([word for word in text.lower().split() if word not in stop_words])
text = stripNonAlphaNum(text)
word_count = {}
```

```python
for word in text:
    try: 
        word_count[word]+=1
    except: 
        word_count[word]=1
```

```python
final_word = []
for i in range(len(word_count)):
    if (word_count.values()[i] > 100):
        final_word.append(word_count.keys()[i])
    else:
        continue
```
